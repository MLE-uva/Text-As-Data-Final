# Create the `TFIDF` table for the Brandon Sanderson Dataset

## Get Data

```python
import pandas as pd
import numpy as np

directory_path  = 'C:/Users/mason/Box/Text As Data Final/Output'
```

```python
DTCM = pd.read_csv(directory_path + "/BrandonSanderson_DTCM.csv", index_col=['title', 'chapter_id'])
DTCM.head()
```

## Define the `TFIDF` function

```python
tf_method = 'sum'
tf_norm_k = .5
idf_method = 'standard'
gradient_cmap = 'YlGnBu'
```

```python
print('TF method:', tf_method)

if tf_method == 'sum':
    TF = DTCM.T / DTCM.T.sum()

elif tf_method == 'smooth':
    TF = (DTCM.T / DTCM.T.sum()) + 1

elif tf_method == 'max':
    TF = DTCM.T / DTCM.T.max()
    
elif tf_method == 'log':
    TF = np.log2(1 + DTCM.T)
    
elif tf_method == 'raw':
    TF = DTCM.T
    
elif tf_method == 'double_norm':
    TF = ((DTCM.T + .5) / (DTCM.T.max() + .5)) + .5
    
elif tf_method == 'binary':
    TF = DTCM.T.astype('bool').astype('int')
    
TF = TF.T
```

```python
DF = DTCM.astype('bool').sum() 
DF
```

```python
N = DTCM.shape[0]
```

```python
print('IDF method:', idf_method)

if idf_method == 'standard':
    IDF = np.log2(N / DF)

elif idf_method == 'max':
    IDF = np.log2(DF.max() / DF) 

elif idf_method == 'plus':
    IDF = np.log2(N / DF) + 1

# This is what SciKit Learn uses
elif idf_method == 'smooth':
    IDF = np.log2((1 + N) / (1 + DF)) + 1
```

```python
TFIDF = TF * IDF
TFIDF.head()
```

## Save output as `TFIDF`

```python
TFIDF.to_csv(directory_path + "/BrandonSanderson_TFIDF.csv")
```
